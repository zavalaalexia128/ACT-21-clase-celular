```python
class Celular:
    def __init__(self, marca, modelo, precio):
        self.marca = marca
        self.modelo = modelo
        self.precio = precio
        self.encendido = False

    def encender(self):
        self.encendido = True
        print(f"El {self.modelo} está encendido.")

    def mostrar_info(self):
        print(f"Celular: {self.marca} {self.modelo}, Precio: ${self.precio}")
#Instanciacion
Cel1 = Celular("Iphone", "Iphone 13", 1000)
Cel2 = Celular("Samsung", "Galaxy S21", 800)
#Llamadas a las funciones para cada objeto celular 1 y celular 2
print("Celular 1:")
Cel1.mostrar_info()
Cel1.encender()

print("\nCelular 2:")
Cel2.mostrar_info()
Cel2.encender()

print("Programa terminado")
print("Programa realizado por Zavala Medina NL= 54")
```
